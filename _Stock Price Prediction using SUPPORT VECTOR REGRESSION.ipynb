## Importing Libraries

```python
import pandas as pd
import numpy as np
```

## Loading the dataset and summarizing the dataset

```python
df = pd.read_csv("StocK price ss.csv")
```

```python
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

## Segregate data into X and Y 

```python
X = df.iloc[:,:-1].values
X
```

```python
Y = df.iloc[:,-1].values
Y
```

## Splitting the dataset in to train test split 

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,Y,test_size=0.25,random_state=0)
```

## Training Dataset using the Support Vector Regression

```python
from sklearn.svm import SVR
model = SVR(coef0=0.20)
model.fit(X_train,y_train)
```

## Prediction for all the data for validation

```python
y_pred = model.predict(X_test)

from sklearn.metrics import r2_score,mean_squared_error
mse = mean_squared_error(y_test,y_pred)
rmse = np.sqrt(mse)
print("Root mean square error:",rmse)
r2score = r2_score(y_test,y_pred)
print("R2Score :",r2score*100)
```
